# Información del Dataset

### Fuente: Kaggle

**Nombre del Dataset:**  US_car_accidents_2016-2021

**Link:**  [Kaggle](https://www.kaggle.com/datasets/sobhanmoosavi/us-accidents)

**Tamaño del dataset:** 1,07 GB

### Descripción:
Este dataset se trata de un conjunto de datos sobre accidentes de tráfico en todo el país, que abarca 49 estados de Estados Unidos. Los datos de accidentes se recopilan desde febrero de 2016 hasta diciembre de 2021, utilizando varias API que proporcionan datos de incidentes (o eventos) de tráfico en tiempo real. Estas API difunden datos de tráfico captados por diversas entidades, como los departamentos de transporte de EE.UU. y estatales, los cuerpos de seguridad, las cámaras de tráfico y los sensores de tráfico dentro de las redes de carreteras. Actualmente, hay unos 2,8 millones de registros de accidentes en este conjunto de datos.

#### Citaciones necesarias:

Como podemos ver en estos últimos casos si sustituimos los valores nan por el valor que más se repite no alteraríamos el valor obtenido y posiblemente ganaríamos más precisión que añadiendo un nuevo caso sólo para tener en cuenta los Nans, por tanto vamos a sustituir:  
**Night -> 1  
Day -> 0  
nan -> valor más repetido en ese caso**

```python
print("Numero de cada valor de Sunrise_Sunset:\n")
# counting the duplicates
dups = df.pivot_table(index = ['Sunrise_Sunset'], aggfunc ='size')
# displaying the duplicate Series
print(dups)

print("Numero de cada valor de Civil_Twilight:\n")
dups = df.pivot_table(index = ['Civil_Twilight'], aggfunc ='size')
print(dups)

print("Numero de cada valor de Nautical_Twilight:\n")
dups = df.pivot_table(index = ['Nautical_Twilight'], aggfunc ='size')
print(dups)

print("Numero de cada valor de Astronomical_Twilight:\n")
dups = df.pivot_table(index = ['Astronomical_Twilight'], aggfunc ='size')
print(dups)
```

Ahora ya tenemos los valores que más se repiten en cada caso con lo cual podemos sustituir el nan por el valor que toque en concreto

# Modelos

<div style="text-align: justify">Inicialmente uno puede pensar que clasificar este dataset mediante un SVM puede ser interesante debido a su base teorica y elevada precisión de clasificación pero este no es el caso. Una clasificación mediante SVM no es adecuada para datasets con una cantidad elevada de datos, esto se debe a que la complejidad de entrenamiento de un SVM es muy elevada debido a su dependencia con la cantidad de datos, por tanto en nuestro caso es mejor evitar este tipo de modelos.</div>

Por esta decisión nuestro modelo inicial va a ser una Red Neuronal Artificial (Artificial Neural Network o ANN). 

## Red Neuronal Artificial (ANN)

<div style="text-align: justify"> Este tipo de algoritmos se basan en el comportamiento de un cerebro animal conformado por millones de neuronas.
Estan conformados por un conjunto de unidades o nodos llamados Neurona, cada conexion entre estas puede transmitir una señal. Esta señal entre las conexiones es un numero real y la salida de cada neurona es calculada mediante una función no lineal de la suma de sus múltiples entradas. Estas entradas suelen contar con cierto peso el cual aumenta o disminuye dependiendo de cómo avance el aprendizaje. Por tanto al tener un conjunto de datos elevados con una gran cantidad de atributos diferentes mediante avance el entrenamiento el propio algoritmo irá restando importáncia a aquellos parámetros poco importantes y aumentándo otros. </div>
