## Conversión valores

Como podemos ver, dentro de todos los atributos que tenemos nos encontramos con 18 de ellos que son de tipo "object" y 13 de ellos que son de tipo "bool". Con tal de poder trabajar con estas columnas vamos a reemplazar sus valores por ints con tal de hacer conversión de valores. Los conjuntos no deberían ser reemplazados, solo adquirir una traduccion entre los tipos.

```python
categorical_columns = df.select_dtypes(include=['object','bool']).columns
categorical_columns
```

```python
from sklearn import preprocessing
#sustituimos las columnas anteriores por ints gracias a sklearn
for i in categorical_columns:
    en = preprocessing.LabelEncoder()
    df[i]= en.fit_transform(df[i])
```

```python
df.head(7)
```

```python
df.info()
```

Como podemos observar ahora contamos con nuestros atributos los cuales eran "object" i "bool" en tipo "int". ¿Cuantos Nans nos siguen quedando en cada caso?

```python
df.isnull().sum()
```

Como podemos observar aquellos atributos boleanos que tenían valores Nan ahora han sido sustituidos por 0 así que puede que se altere alguno de los valores, pero cómo teníamos relativamente pocos valores Nan, la pérdida de información será pequeña.  
Los valores que no han sido rellenados aún ya que el tipo que tenían no ha variado, por lo tanto en estos casos lo que haremos será sustituir estos valores nulos por la media de los valores que tenían, con tal de no alterar las posibles salidas.

```python
df['Humidity(%)'] = df['Humidity(%)'].fillna(df['Humidity(%)'].mean())
df['Pressure(in)'] = df['Pressure(in)'].fillna(df['Pressure(in)'].mean())
df['Visibility(mi)'] = df['Visibility(mi)'].fillna(df['Visibility(mi)'].mean())
print(df.shape)
print("\nNan en la nueva base de datos:\n",df.isnull().sum(), "\n")
```

Una vez hemos conseguido tratar todos los datos, podemos empezar a entrenar, pero antes de ello, vamos a observar que atributos son los más importantes

## Atributos de mayor importáncia

```python
train_y = df['Severity'].values
x_cols = [col for col in df.columns if col not in ['Severity'] if df[col].dtype=='float64'or df[col].dtype=='int64']
train_col= df[x_cols]

fearture_name = train_col.columns.values 

from sklearn import ensemble 

model = ensemble.ExtraTreesRegressor(n_estimators=25, max_depth=30, max_features=0.3, n_jobs=-1, random_state=0)
model.fit(train_col,train_y)


importance = model.feature_importances_
std = np.std([tree.feature_importances_ for tree in model.estimators_],axis=0)
indices = np.argsort(importance)[::-1][:20]

plt.figure(figsize=(20,7))
plt.title("Importancia Atributos")
plt.bar(range(len(indices)), importance[indices], color="r", yerr=std[indices], align="center")
plt.xticks(range(len(indices)), fearture_name[indices], rotation='vertical')
plt.xlim([-1, len(indices)])
plt.show()
```

Seguidamente observamos los resultados obtenidos con un heatmap para ver si los resultados parecen correctos.

```python
df_corr = df.corr()
plt.figure(figsize=(25,25))
ax = sns.heatmap(df_corr,cmap='RdYlGn',annot=True,center=0,vmin=-1.0,vmax=1.0)
plt.show()
```

Como podemos observar, nos encontramos con ciertos casos en los que la correlacion entre los valores es de 1 aún y no ser un atributo con él mismo. Esto se debe muy posiblemente a que los valores en estos casos son iguales. Por ejemplo que el valor de la Latitud en el inicio sea el mismo valor que la Latitud al finalizar.
También encontramos el caso de valores con correlaciones inexactas derivados de la substitución de su valor como una variable de tipo "object" a una de tipo "int".
